## <a name="objetivo"> 1. Objetivo <a>

El objetivo de este informe se caracteriza por analizar el progreso de las ventas de las 10 tiendas de la cadena de supermercados TOSCOS durante los años 2013-2015, con la motivación de estudiar como se han comportado los locales según las diferentes zonas. Adicionalmente, se presentarán los pronósticos de ventas para el periodo comprendido entre el 1 de agosto y el 10 de septiembre de 2015.

## <a name="metodologia"> 2. Metodología <a>

Para la realización del estudio se ha procedido con un análisis estadístico de las diferentes series temporales de cada local, con el objetivo de detectar sus características y seleccionar los modelos de predicción que mejor se ajustaran a ellas. Se han utilizado diferentes librerías estadísticas del lenguaje Python. 

## <a name="resultados"> 3. Resultados <a>

En cuanto al análisis de la situación actual de los supermercados, procederemos a observar una captura de las ventas globales con el objetivo de conocer cómo se han comportado, una vez realizado el análisis se procederá a seleccionar las 3 zonas donde operan nuestros locales agrupando las ventas semanales, teniendo en cuenta el nº de locales: zona 1(3 locales), zona 2(3 locales), zona 3 (4 locales).

**Cabe destacar que respecto a las tiendas individuales las que mayor nº de ventas han realizado han sido T1b,T2a y T3b**

### 3.1. Ventas totales

La visión global del nº de ventas obtenidas en todos los locales de la compañía es bastante positivo, dado que observamos como han tenido una tendencia positiva a largo plazo, observamos como la media durante el año 2013 fue de aproximadamente 375.000 ventas semanales, en 2014 la media disminuyó debido principalmente a una caida durante la segunda mitad de año. 

Sin embargo, durante el año 2015, logramos aumentar las ventas manteniendolas en una media semanal de aproximadamente unos 430.000, lo que supone un aumento del 14%, lo cual es bastante positivo. 

Conocer el nº de ventas global es interesante, sin embargo, vamos a desgranar el nº de ventas semanales por zona, con el objetivo de ver como se han ido comportando, si han sido similares o si existen diferencias significativas a tener en cuenta entre las diferentes zonas operativas.


### 3.2. Zona 1

Como podemos observar en la gráfica las ventas de la zona 1 han ido incrementando semanalmente. Comenzamos en 2013 donde estuvimos oscilando entre las 80.000 ventas por semana hasta mediados de año, donde se produce un aumento significativo que se mantuvo constante hasta julio de 2015, con una media entorno a las 110.000 ventas aprox. Es decir, la zona 1 ha incrementado el nº de ventas por un 37% aproximadamente. 

Cabe destacar como se mantiene una tendencia a finales de año que dispara las ventas por encima de lo normal, probablemente dadas las fechas festivas de las Navidades provocando un aumento de la demanda.


### 3.2. Zona 2

En cuanto respecta a la zona 2 observamos como al comienzo de 2013 teníamos una media de 110.000 ventas, esta a lo largo del tiempo va incrementando, llegando a julio de 2015 con una media de 135.000 ventas, es decir un 22% de crecimiento aproximado. 
Como observamos la tendencia es creciente y existen movimientos irregulares a finales de año y mediados de 2013.

Los de final de año se corresponde por la navidad, tal y como ocurría en la zona 1, sin embargo, nos encontramos con que en el año 2013 hay un movimiento atípico en el nº de ventas, del cual habría que estudiar las causas que lo provocaron.


### 3.3. Zona 3

En cuanto lo que respecta a la zona 3, observamos como desde comienzos de 2013 las ventas fueron creciendo ligeramente hasta mediados de 2014. En julio de ese año las ventas cayeron drásticamente manteniéndose bastante bajas los 6 meses siguientes, y no es hasta comienzos de 2015, que las ventas vuelven a incrementar drásticamente retomando la senda creciente de los primeros años.


En esta zona se repiten los movimientos navideños, sin embargo, sería interesante contar con todos los datos que nos permitan conocer que ocurrió durante el periodo comprendido entere junio y diciembre de 2014 para entender los motivos y evitar que se repitan en el futuro.
