## <a name="temporal"> 2. Resultados del análisis de la serie temporal</a>

### 2.1. Histórico de los ingresos de la compañia

La serie temporal de Apple, como observamos en la gráfica inferior, nos muestra un incremento de los ingresos a lo largo de los años desde su comienzo. Sin embargo se puede apreciar como a partir del **2007-2008 su valor comienza a incrementar en mayor proporcion que durante los años 1990-2006**.

Esto se debe a que **a partir del 2007**, con el lanzamiento del primer iPhone, **revolucionan el mercado** de dispositivos móviles y comienzan a sacar nuevos productos como el iPad, MacBook o Apple Watch, además de  nuevos servicios monetizados en la nube.
Todo esto **permitió lograr  un notable incremento de los ingresos, reputación y visibilidad dentro del mercado**.  

### 2.2. Estacionalidad en media o varianza

Decimos que una serie es estacional en media y varianza cuando la media y la variabilidad se mantienen constantes a lo largo del tiempo y además, no presenta tendencia.

Una vez descompuesta la serie temporal observamos que **NO es estacional en media**  ya que presenta una tendencia claramente alcista a lo largo de los años y **NO es estacional en varianza**, dado que no se mantiene constatnte a lo largo de los años. Sin embargo observamos un componente estacional, que analizaremos en más detalle en el siguiente apartado.

### 2.3. Componente estacional 

En cuanto al componente estacional, observamos como su **comportamiento es muy similar entre los trimestres**. Todos con un patrón **alcista y muy pronunciado**. En cuanto a la media de ingresos observamos como el **primer trimestre fiscal es superior**. Esto se puede deber a que durante los meses de **octubre y noviembre, la compañía suele sacar al mercado las nuevas generaciones de sus productos potenciales**, tanto hardware (iPhone, iPad, Macbook y Apple Watch) como software (Nueva versión de IOS y nuevos servicios en la nube).

Además a todo esto se le junta que **durante este periodo existe un gran número de campañas comerciales** como son el día del soltero (11/11, especialmente en China), Black Friday (25/11) y el periodo de Navidad (Diciembre).

##  <a name="covid"> 5. Efecto COVID </a>

Dado que el COVID surgió a finales de 2020 en China, tendremos en cuenta que **afectó a la compañia a partir del segundo trimestre fiscal de 2020 (Febrero- Marzo)** dado que mundialmente la gran parte de confinamientos surgieron a partir de esas fechas.

Es por ello que para el cálculo del efecto lo realizaremos **hasta el 1er trimestre de 2021**, que es cuando empezaron a administrar las primeras dosis de la vacuna y al ver su buen funcioanmiento, flexibilizar las medidas y lograr la 'normalidad'.

En la tabla observamos como el efecto COVID tuvo un **impacto muy positivo** para la empresa llegando a ingresar en sus arcas unos 29.4 billones de dolares superando todas las previsiones. Esto se debe a que en los momentos donde la gente no podía salir de casa o tenian movilidad restringida, surge la necesidad de entretenimiento y contacto social, es por ello que los productos de Apple tuvieron un incremento de las ventas, lo que se traduce en un incremento de sus beneficios.

## <a name="comparacion"> 7. Comparación de las predicciones del modelo ARIMA con las de Barclays </a>

Respecto a las predicciones del modelo ARIMA frente a las de Barclays, observamos que las del año 2021 son muy similares con **tan solo un 0,92% de diferencia en el año 2021**, por lo que es bastante ajustado y creible.

Sin emabrgo, para el año 2022, nuestro modelo ARIMA estima unos ingresos bastante más superiores que los de Barclays, con una **variacion del 15%**, por lo que habría que estudiar que ha tenido en cuenta Barclays, que nuestro modelo no, para poder detectar posibles fallos.

## <a name="conclusiones"> 8. Conclusiones Generales. </a>

- La **tendencia de la compañía ha sido muy alcista** durante toda la historia de la empresa, pero como comentamos anteriormente, especialmente desde que lanzó al mercado productos innovadores como el iPhone, iPad, etc.

- La serie temporal de Apple hemos comprobado que **no es estacional en media ni en varianza**, sin embargo, **si que tiene un  componente estacional** que se suele repetir anualmente, con mayor incidencia el primer trimestre fiscal.

- El número de **ventas de la compañia ha tenido variaciones muy positivas** a lo largo de los años, llegando a superar en algunos casos el 60%  entre año y año.

- Durante la estimación de modelos de predicción, el que **mejor resultado nos ha dado ha sido el modelo ARIMA**, ya que reflejaba un menor error que el modelo ETS. Por lo que le hemos utilizado para predecir lo que quedaba del año 2021 y 2022.

- Tras comprobar un posible **efecto COVID**, podemos afirmar que si que lo hubo, pero **afectando favorablemente a la empresa**, dado que incrementó sus ventas de una manera significativa, superando las predicciones estimadas.
